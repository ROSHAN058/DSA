# Two Sum

## Problem Statement

Given an array of integers `nums` and an integer `target`, return the indices of the two numbers that add up to the target.

### Constraints
- Each input contains exactly one valid solution.
- You may not use the same element twice.
- The answer can be returned in any order.

### Example
- Input: `nums = [2, 7, 11, 15]`, `target = 9`
- Output: `[0, 1]`

### Explanation
Because `nums[0] + nums[1] == 9`, the correct result is `[0, 1]`.

```python
def two_sum(nums, target):
    seen = set()
    for i, num in enumerate(nums):
        complement = target - num
        if complement in seen:
            return [nums.index(complement), i]
        seen.add(num)
    return []

nums = [3, 3]
target = 6
result = two_sum(nums, target)
print(result)
```

# Merge Two Sorted Arrays

Merge two sorted arrays into a single sorted array while preserving order.

Example
- Input: `nums1 = [1, 3, 5]`, `nums2 = [2, 4, 6]`
- Output: `[1, 2, 3, 4, 5, 6]`

Approach
1. Initialize two pointers, one for each array.
2. Compare the current elements at the pointers.
3. Append the smaller element to the merged result.
4. Advance the pointer for the array from which the element was taken.
5. When one array is exhausted, append the remaining elements from the other array.

```python
nums1 = [1, 3, 5]
nums2 = [2, 4, 6]
#merge two sorted arrays code
merged = []
i, j = 0, 0

while i < len(nums1) and j < len(nums2):
    if nums1[i] < nums2[j]:
        merged.append(nums1[i])
        i += 1
    else:
        merged.append(nums2[j])
        j += 1

# Append any remaining elements from either array
merged.extend(nums1[i:])
merged.extend(nums2[j:])

print(merged)
```
