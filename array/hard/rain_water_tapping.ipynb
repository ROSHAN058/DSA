# Trapping Rain Water

## Problem Statement
Given an array of non-negative integers representing an elevation map, where each element is a bar of width 1, determine how much water can be trapped after raining.

### Example
- Input: `height = [0, 1, 0, 2, 1, 0, 1, 3, 2, 1, 2, 1]`
- Output: `6`

### Explanation
The elevation map can trap 6 units of water.

### Visual Idea
A simple way to imagine this is water being trapped between higher bars:

![Rain water trapping illustration](rain_water_example.png)


```python
def trap(height):
    left, right = 0, len(height) - 1
    left_max = 0
    right_max = 0
    trapped = 0

    while left < right:
        if height[left] <= height[right]:
            if height[left] >= left_max:
                left_max = height[left]
            else:
                trapped += left_max - height[left]
            left += 1
        else:
            if height[right] >= right_max:
                right_max = height[right]
            else:
                trapped += right_max - height[right]
            right -= 1

    return trapped

height = [0, 1, 0, 2, 1, 0, 1, 3, 2, 1, 2, 1]
print(trap(height))
```

### Alternative Approach: Prefix and Suffix Arrays
This method first finds the tallest bar to the left of each position and the tallest bar to the right. At each index, the water level is limited by the smaller of those two values. If the current bar is shorter than that limit, the difference is the trapped water at that position.

```python
def trap_with_prefix_suffix(height):
    if not height:
        return 0

    n = len(height)
    left_max = [0] * n
    right_max = [0] * n

    left_max[0] = height[0]
    for i in range(1, n):
        left_max[i] = max(left_max[i - 1], height[i])

    right_max[-1] = height[-1]
    for i in range(n - 2, -1, -1):
        right_max[i] = max(right_max[i + 1], height[i])
    print(left_max, right_max)
    return sum(min(left_max[i], right_max[i]) - height[i] for i in range(n))

height = [0, 1, 0, 2, 1, 0, 1, 3, 2, 1, 2, 1]
print(trap_with_prefix_suffix(height))
```
