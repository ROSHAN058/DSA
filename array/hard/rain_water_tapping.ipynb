# Trapping Rain Water

## Problem Statement
Given an array of non-negative integers representing an elevation map, where each element is a bar of width 1, determine how much water can be trapped after raining.

### Example
- Input: `height = [0, 1, 0, 2, 1, 0, 1, 3, 2, 1, 2, 1]`
- Output: `6`

### Explanation
The elevation map can trap 6 units of water.

### Visual Idea
A simple way to imagine this is water being trapped between higher bars:

![Rain water trapping illustration](https://upload.wikimedia.org/wikipedia/commons/8/8c/Container_Example.jpg)


```python
def trap(height):
    left, right = 0, len(height) - 1
    left_max = 0
    right_max = 0
    trapped = 0

    while left < right:
        if height[left] <= height[right]:
            if height[left] >= left_max:
                left_max = height[left]
            else:
                trapped += left_max - height[left]
            left += 1
        else:
            if height[right] >= right_max:
                right_max = height[right]
            else:
                trapped += right_max - height[right]
            right -= 1

    return trapped

height = [0, 1, 0, 2, 1, 0, 1, 3, 2, 1, 2, 1]
print(trap(height))
```
